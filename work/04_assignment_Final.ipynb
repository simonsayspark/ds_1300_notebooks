```python

import matplotlib.animation as animation
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
print("For this investigation I have decided to compare a data set that shows how the response variable: Log of NO2, is affected by the variables: Log of cars, Log of temperature, and log of windspeed")
```

```python

raw_df = pd.read_csv("../data/NO2.dat.txt", sep="\s+", header=None)
data = raw_df.values[:, 1] # remove data from the first column, and start from the second column (Log of cars)
target = raw_df.values[:, 0] # independent varibale, the NO2 we want to predict based on other variabels

mask = raw_df.values[:,0].mean() 


aboveMean = target > mask 
belowMean = target < mask  


plt.scatter(data[aboveMean], target[aboveMean], alpha=0.4, color="cyan", label="Above Mean")
plt.scatter(data[belowMean], target[belowMean], alpha=0.4, color="magenta", label="Below Mean")

plt.title("Log of cars vs log of N02")
plt.xlabel('Log of the number of cars')
plt.ylabel('Log of N02 concetration')
plt.legend()
plt.show()
```

```python
print("Description:")
print("The scatter plot shows the relation between the log of the number of cars in relation to the log of the concentration of NO2. The legend describes the values above the mean of NO2 concentration above and below.")

print("")
print("Method of plotting that I chose:") 
print("I chose to make it a scatter plot because it looks nicer than other kinds of plots. Moreover, I decided to calculate the mean, because I wanted to see values above and below the mean of the NO2 concentration.")

```

```python
data = raw_df.values[:, 2] # remove data from the first column, and start from the second column (Log of cars)
target = raw_df.values[:, 0] # independent varibale, the NO2 we want to predict based on other variabels

mask = raw_df.values[:,0].mean()


aboveMean = target > mask
belowMean = target < mask

plt.scatter(data[aboveMean], target[aboveMean], alpha=0.4, color ="red" , label = "Above Mean" )
plt.scatter(data[belowMean], target[belowMean], alpha = 0.4, color = "blue", label = "Below Mean")

plt.title("Log of temp vs log of N02")
plt.xlabel('Log of the Temperature')
plt.ylabel('Log of N02 concetration')
plt.legend()
plt.show()
```

```python
print("Description:")
print("The scatter plot shows the relation between the log of temperature in relation to the log of the concentration of NO2.")


```
