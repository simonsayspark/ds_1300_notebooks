```python

def lin_reg(x,y):
    # Calculate the average x and y
    x_avg = np.mean(x)
    y_avg = np.mean(y)

    num = 0
    den = 0
    for i in range(len(x)): # This represents our sums
        num = num + (x[i] - x_avg)*(y[i] - y_avg) # Our numerator
        den = den + (x[i] - x_avg)**2 # Our denominator
    # Calculate slope
    m = num / den
    # Calculate intercept
    b = y_avg - m*x_avg

    print (m, b)
    
    # Calculate our predicted y values
    y_pred = m*x + b
    
    return y_pred

```

```python
X= concatData['DateInt'].values
Y= concatData["Result\nValue\n(Def.)"].values                
Y_pred = lin_reg(X,Y)
                                   
concatData.plot.scatter(x = "Date" , y = "Result\nValue\n(Def.)", rot =45, figsize =(20,10))
plt.plot([min(concatData['Date'].values), max(concatData['Date'].values)], [min(Y_pred), max(Y_pred)], color='red')
plt.title("East Coast Data Set")
```
